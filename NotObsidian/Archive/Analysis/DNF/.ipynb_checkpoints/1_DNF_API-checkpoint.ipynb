```python
import sys
import requests
from urllib import parse

sys.stdin = open('../dnf_api_key.txt', 'r')


api_key = sys.stdin.readline().strip()

# 한글이나 특수 문자를 URL에 보낼때는 URL인코딩이 필요함
# URL 인코딩 = 영어, 숫자 등을 제외한 나머지 문자는 1바이트 단위의 16진수로

gateway = f"https://api.neople.co.kr/df/"

# 캐릭터 검색
characterName = parse.quote('回歸')
character = f"servers/cain/characters?characterName={characterName}&apikey={api_key}"




# 캐릭터 기본 정보 검색
def search_info(character_id):
    return f"https://api.neople.co.kr/df/servers/cain/characters/{character_id}?apikey={api_key}"

def search_status(character_id):
    return f"https://api.neople.co.kr/df/servers/cain/characters/{character_id}/status?apikey={api_key}"

def search_timeline(character_id):
    # 디폴트 : start = 오늘부터 30일 전 / end = 오늘
    return f"https://api.neople.co.kr/df/servers/cain/characters/{character_id}/timeline?&code=401,402,403,404,405,406,505,511&apikey={api_key}"

url = gateway + character
r = requests.get(url)
```

```python
import pandas as pd
print(r)
print(r.json())
df = pd.DataFrame(r.json()['rows'])

url = search_status(df['characterId'][0])
r = requests.get(url)
r.json()

url2 = search_timeline(df['characterId'][0])
r = requests.get(url2)
r.json()
```
