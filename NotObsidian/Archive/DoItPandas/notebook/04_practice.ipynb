# 앤스콤 데이터 집합 불러오기

```python
import seaborn as sns

anscombe = sns.load_dataset("anscombe")
print(anscombe)
print(type(anscombe))
```

```python
print(anscombe[anscombe['dataset'] == 'I'].describe())
print(anscombe[anscombe['dataset'] == 'II'].describe())
print(anscombe[anscombe['dataset'] == 'III'].describe())
anscombe[anscombe['dataset'] == 'IV'].describe()
```

# matplotlib 라이브러리로 간단한 그래프 그리기

```python
%matplotlib inline
import matplotlib.pyplot as plt

dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']
dataset_3 = anscombe[anscombe['dataset'] == 'III']
dataset_4 = anscombe[anscombe['dataset'] == 'IV']
```

```python
fig, ax = plt.subplots(224)

ax[0][0].plot(dataset_1['x'], dataset_1['y'], 'o') # 마지막 argument 'o'는 점으로 그린다는 뜻
ax[0][1].plot(dataset_2['x'], dataset_2['y'], 'o') # 마지막 argument 'o'는 점으로 그린다는 뜻
ax[1][0].plot(dataset_3['x'], dataset_3['y'], 'o') # 마지막 argument 'o'는 점으로 그린다는 뜻
ax[1][1].plot(dataset_4['x'], dataset_4['y'], 'o') # 마지막 argument 'o'는 점으로 그린다는 뜻

plt.show()
```

# 한 번에 4개의 그래프 그리기

```python
# 책은 이렇게 분할함

fig = plt.figure()

ax1= fig.add_subplot(2, 2, 1)
ax2= fig.add_subplot(2, 2, 2)
ax3= fig.add_subplot(2, 2, 3)
ax4= fig.add_subplot(2, 2, 4)

ax1.plot(dataset_1['x'], dataset_1['y'], 'o')
ax2.plot(dataset_2['x'], dataset_2['y'], 'o')
ax3.plot(dataset_3['x'], dataset_3['y'], 'o')
ax4.plot(dataset_4['x'], dataset_4['y'], 'o')

plt.tight_layout
```

```python
fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(dataset_1['x'], dataset_1['y'], 'o')
ax[0, 1].plot(dataset_2['x'], dataset_2['y'], 'o')
ax[1, 0].plot(dataset_3['x'], dataset_3['y'], 'o')
ax[1, 1].plot(dataset_4['x'], dataset_4['y'], 'o')
```

# 기초 그래프 그리기 - 히스토그램, 산점도, 박스 그래프

```python
import seaborn as sns
tips = sns.load_dataset('tips')
print(tips.head())
print(type(tips))
```

# 단변량 그래프 그리기 - 히스토그램

```python
fig = plt.figure()
ax1 = fig.add_subplot(1, 1, 1)
ax1.hist(tips['total_bill'], bins = 10)
ax1.set_title('Histrogram of Total Bill')
ax1.set_xlabel('Frequency')
ax1.set_ylabel('total bill')
```

# 다변량 그래프 그리기

# 데이터프레임과 시리즈로 그래프 그리기

# 알아두면 좋아요

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug = True)
```

```python
sns.regplot(x='total_bill', 
           y= 'tip',
           data = tips,
           fit_reg = True)
```

```python
joint = sns.jointplot(x = 'total_bill',
                     y = 'tip',
                     data = tips)
```

```python
joint = sns.jointplot(x = 'total_bill',
                     y = 'tip',
                     data = tips,
                     kind = "hex")
```

```python
ax = sns.kdeplot(data = tips['total_bill'],
                data2 = tips['tip'],
                shade = True)
```

```python
sns.barplot(x = 'time', y = 'total_bill', data = tips)
```

```python
# 박스 그래프
sns.boxplot(x = 'time',
           y = 'total_bill',
           data = tips)
```

```python
sns.violinplot(x = 'time',
           y = 'total_bill',
           data = tips)
```

```python
sns.pairplot(tips)
```

```python
# 정보 중첩 피하기 : 위치 지정해서 직접 그리기
pair_grid = sns.PairGrid(tips)
pair_grid = pair_grid.map_upper(sns.regplot)
pair_grid = pair_grid.map_lower(sns.kdeplot)
pair_grid = pair_grid.map_diag(sns.distplot, rug = True)
plt.show()
```
