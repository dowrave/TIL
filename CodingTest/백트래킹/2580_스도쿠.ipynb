```python
import sys
sys.stdin = open('test.txt', 'r')
input = sys.stdin.readline
matrix = []
for i in range(9):
    matrix.append(list(map(int, input().split())))

empty_lst = []
for j in range(9):
    for k in range(9):
        if matrix[j][k] == 0:
            empty_lst.append([j, k])
```

```python
# 일단 조건식부터 써보자

# 1. 어떤 범위 내에서 없는 값 찾기
def find_conditions(a, b):
    num_set = set([i for i in range(1, 10)])
    
    def find_rows(a, b):
        target = matrix[a]
        not_in = list(num_set - set(target))
        
        if len(not_in) == 1:
            matrix[a][b] = not_in[0]
            return True


    def find_cols(a, b):
        target = [rows[b] for rows in matrix]
        not_in = list(num_set - set(target))
        if len(not_in) == 1:
            matrix[a][b] = not_in[0]
            return True


    def find_squares(a, b):
        # [n // 3 : n // 3 + 3] 아닌가?    
        target = sum([rows[b // 3 * 3: b // 3 * 3+ 3] for rows in matrix[a // 3 * 3: a // 3 * 3+ 3]], [])
        not_in = list(num_set - set(target))
        if len(not_in) == 1:
            matrix[a][b] = not_in[0]
            return True
    
    if find_rows(a,b):
        return True
    if find_cols(a,b):
        return True
    if find_squares(a,b):
        return True
    
# 2. dfs 구현 - n은 empty_lst의 인덱스
def dfs(n):
    if n == len(empty_lst):
        for i in matrix:
            print(" ".join(map(str, i)))
        return
    
    a = empty_lst[n][0]
    b = empty_lst[n][1]

    if find_conditions(a, b):

        dfs(n+1)
        matrix[a][b] = 0

dfs(0)
```

    - 그리고 검사하는 함수는 검사만 하는 게 좋다
    - 값을 넣는 건 함수 밖에서 구현하는 게 좋다
    - 그렇게 하기 위해 반복문을 쓰는 것 : 넣을 값에 대해 3가지 조건이 다 충족하는지 보고, 충족한다면 그 값을 넣고 다음 재귀식으로 넘어가는 방식

```python
import sys

# -----------------------------------------------------------------------------------------------
def sudoku(idx):
    
    global finished
    
    if idx == len(blank_lst):
        for line in map_lst:
            print(" ".join(map(str, line)))
#             print(*line)
            finished = True
            
    else:
        x = blank_lst[idx][0]
        y = blank_lst[idx][1]

        for value in range(1, 10):
            if cond_row(x, value) and cond_col(y, value) and cond_square(x, y, value):
                map_lst[x][y] = value
                sudoku(idx + 1)
                map_lst[x][y] = 0
                if finished == True:
                    break
                    
# 조건 3개
def cond_row(x, val):
    for i in range(9):
        if val == map_lst[x][i]:
            return False
    return True

def cond_col(y, val):
    for i in range(9):
        if val == map_lst[i][y]:
            return False
    return True

def cond_square(x, y, val):
    nx = x // 3 * 3
    ny = y // 3 * 3
    
    for i in range(0, 3):
        for j in range(0, 3):
            if val == map_lst[nx+i][ny+j]:
                return False
    return True
# ----------------------------------------------------------------------------------

input = sys.stdin.readline

map_lst = []
for i in range(9):
    map_lst.append(list(map(int, input().split())))

# blank 찾기
blank_lst = []
for i in range(9):
    for j in range(9):
        if map_lst[i][j] == 0:
            blank_lst.append((i, j))

finished = False
sudoku(0)
```
