```python
import sys
sys.stdin = open('test.txt', 'r')
T = int(sys.stdin.readline())

for i in range(T):
    x1, y1, r1, x2, y2, r2 = map(int, sys.stdin.readline().split())
    

    
    d = ((x2-x1)**2 + (y2-y1)**2)**(1/2)

    if d >= r1 and d >= r2: # 두 점은 서로 원의 중심보다 밖에 있음
        if r1 + r2 > d:
            print(2)

        elif r1 + r2 == d:
            print(1)

        else:
            print(0)

    else: # 위의 경우를 제외한 나머지 
        if x1 == x2 and y1 == y2 and r1 == r2:
            print(-1)

        elif abs(r1 - r2) < d:
            print(2)
            
        elif abs(r1 - r2) == d:
            print(1)
        
        else:

            print(0)
```
